```python


'''
n07772935: coconut 
 
n02124075: Egyptian cat 
 
n01739381: vine snake

'''
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Function to display random images from a directory
def display_random_images(directory, category, num_images=10):
    plt.figure(figsize=(15, 8))
    plt.suptitle(f"Random Images of {category.capitalize()}", fontsize=16)
    
    # List all files in the directory
    files = os.listdir(directory)
    # Randomly select num_images
    selected_files = random.sample(files, num_images)
    
    for i, file_name in enumerate(selected_files, 1):
        # Open and display the image
        img = Image.open(os.path.join(directory, file_name))
        plt.subplot(2, 5, i)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Image {i}")
    
    plt.show()

# Directory paths for each category
coconut_directory = 'D:\\Data_Exploration\\Images\\n07772935'
vine_snake_directory = 'D:\\Data_Exploration\\Images\\n01739381'
egyptian_cat_directory = 'D:\\Data_Exploration\\Images\\n02124075'

# Display 10 random images from each category
display_random_images(coconut_directory, "coconut")
display_random_images(vine_snake_directory, "vine snake")
display_random_images(egyptian_cat_directory, "egyptian cat")
```
